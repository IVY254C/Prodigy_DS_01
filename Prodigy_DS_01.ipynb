## Extarction of top 10 countries with respect to total population

```python
# Filter data for total population based on 'Country Name' or 'Country Code' as needed

# Sort data based on the total population for 2023
total_population_sorted = df.sort_values(by="2023", ascending=False)

# Get the top ten countries with the highest total population for 2022
total_top_ten_countries = total_population_sorted.head(10)

# Display the top 10 countries with their names and codes
print("Top ten countries by total population in 2023\n")
print(total_top_ten_countries[['Country Name', 'Country Code', '2023']])
```

# Bar Plot


## Top ten countries of total population in year 2022 and 2016

```python
# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2023", y="Country Code", data=total_top_ten_countries)
plt.title("Top Ten Countries of Total Population (2023)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022", y="Country Code", data=total_top_ten_countries)
plt.title("Top Ten Countries of Total Population (2022)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Extraction of bottom 10 countries with respect to total popultaion

```python
# Sort data in ascending order based on total population for 2023
total_population_sorted = df.sort_values(by="2023", ascending=True)

# Get the bottom 10 countries with the lowest total population
bottom_ten_countries = total_population_sorted.head(10)

# Display results
print("Bottom ten countries by total population in 2023:\n")
print(bottom_ten_countries[["Country Name", "Country Code", "2023"]])

```

## Bottom ten countries of total population in year 2022 and 2023

```python
import matplotlib.pyplot as plt

# Sort data in ascending order based on total population for 2022
total_population_sorted_2022 = df.sort_values(by="2022", ascending=True)

# Get the bottom 10 countries with the lowest total population in 2022
bottom_ten_countries_2022 = total_population_sorted_2022.head(10)

# Create a bar plot for the bottom 10 countries in 2022
plt.figure(figsize=(16,8))
plt.bar(bottom_ten_countries_2022['Country Name'], bottom_ten_countries_2022['2022'], color='skyblue')
plt.title('Bottom 10 Countries by Total Population in 2022')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha="right")  # Rotate country names for readability
plt.tight_layout()  # Adjust layout for better fitting
plt.show()

# Sort data in ascending order based on total population for 2023
total_population_sorted_2023 = df.sort_values(by="2023", ascending=True)

# Get the bottom 10 countries with the lowest total population in 2023
bottom_ten_countries_2023 = total_population_sorted_2023.head(10)

# Create a bar plot for the bottom 10 countries in 2023
plt.figure(figsize=(16,8))
plt.bar(bottom_ten_countries_2023['Country Name'], bottom_ten_countries_2023['2023'], color='lightgreen')
plt.title('Bottom 10 Countries by Total Population in 2023')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha="right")  # Rotate country names for readability
plt.tight_layout()  # Adjust layout for better fitting
plt.show()
```
